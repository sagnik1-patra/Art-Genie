```python
# =========================
# ArtGenie: Predict + Accuracy Graph + Confusion Heatmap (robust to H5 TextVectorization bug)
# =========================

import os, re, json, logging
from typing import List, Optional
import numpy as np, pandas as pd, matplotlib.pyplot as plt, joblib
import tensorflow as tf
from tensorflow.keras.models import load_model

INPUT_CSV   = r"C:\Users\sagni\Downloads\Art Genie\archive\wikiart_scraped.csv"
OUTPUT_CSV  = r"C:\Users\sagni\Downloads\Art Genie\predictions.csv"
ARTIFACT_DIR= r"C:\Users\sagni\Downloads\Art Genie"

PKL_PATH       = os.path.join(ARTIFACT_DIR, "artgenie_textclf.pkl")
H5_PATH        = os.path.join(ARTIFACT_DIR, "artgenie_textclf.h5")    # may fail due to TextVectorization table init
KERAS_PATH_ALT = os.path.join(ARTIFACT_DIR, "artgenie_textclf.keras") # preferred SavedModel/KerasV3 format if you re-save
HISTORY_CSV    = os.path.join(ARTIFACT_DIR, "history.csv")
ACC_PNG        = os.path.join(ARTIFACT_DIR, "accuracy_curve.png")
CONF_FULL_CSV  = os.path.join(ARTIFACT_DIR, "confusion_matrix_full.csv")
HEATMAP_PNG    = os.path.join(ARTIFACT_DIR, "confusion_heatmap_top25.png")

logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)s | %(message)s")

POSSIBLE_TEXT_COLS = ["title","description","caption","tags","genre","style","artist","movement","content","about","wiki","text","meta","materials","subject"]
POSSIBLE_LABEL_COLS = ["style","genre","artist","label"]

def normalize_cols(df): 
    df = df.copy(); df.columns=[str(c).strip().lower().replace(" ","_") for c in df.columns]; return df
def canonicalize(s): return re.sub(r"\s+"," ",str(s).lower()).strip()
def collect_text_columns(df):
    cols=[c for c in POSSIBLE_TEXT_COLS if c in df.columns]
    if not cols: cols=[c for c in df.columns if df[c].dtype==object]
    return cols[:8]
def build_text_series(df, cols):
    parts=[df[c].astype(str) for c in cols]; txt=parts[0]
    for p in parts[1:]: txt=txt.str.cat(p, sep=" . ", na_rep="")
    return txt.fillna("").map(lambda s: re.sub(r"\s+"," ",str(s)).strip())
def pick_label_column(df):
    for c in POSSIBLE_LABEL_COLS:
        if c in df.columns and df[c].notna().any(): return c
    return None

def plot_accuracy_curve(history_csv, out_png, also_show=True):
    if not os.path.exists(history_csv):
        logging.warning(f"[PLOT] history.csv not found at {history_csv} — skipping accuracy plot."); return
    hist=pd.read_csv(history_csv); plt.figure(figsize=(9,6))
    if "accuracy" in hist.columns: plt.plot(hist["epoch"], hist["accuracy"], label="Train Accuracy")
    if "val_accuracy" in hist.columns: plt.plot(hist["epoch"], hist["val_accuracy"], label="Val Accuracy")
    if "top3_acc" in hist.columns: plt.plot(hist["epoch"], hist["top3_acc"], label="Train Top-3")
    if "val_top3_acc" in hist.columns: plt.plot(hist["epoch"], hist["val_top3_acc"], label="Val Top-3")
    plt.title("Training Accuracy"); plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend(); plt.grid(True, linestyle="--", linewidth=0.5)
    plt.tight_layout(); plt.savefig(out_png, dpi=160)
    if also_show:
        try: plt.show()
        except Exception: pass
    plt.close(); logging.info(f"[PLOT] Saved accuracy curve → {out_png}")

def plot_confusion_heatmap(y_true_idx, y_pred_idx, class_names, out_png, full_csv, top_k=25, also_show=True):
    from sklearn.metrics import confusion_matrix
    cm_full=confusion_matrix(y_true_idx, y_pred_idx, labels=np.arange(len(class_names)))
    pd.DataFrame(cm_full, index=class_names, columns=class_names).to_csv(full_csv, encoding="utf-8")
    logging.info(f"[PLOT] Saved full confusion matrix CSV → {full_csv}")
    true_counts=cm_full.sum(axis=1); top_idx=np.argsort(true_counts)[::-1][:min(top_k,len(class_names))]
    sub=cm_full[np.ix_(top_idx, top_idx)]; sub_labels=[class_names[i] for i in top_idx]
    plt.figure(figsize=(12,10)); im=plt.imshow(sub, interpolation="nearest", aspect="auto"); plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.title(f"Confusion Matrix Heatmap (Top {len(sub_labels)} Classes)"); plt.xticks(range(len(sub_labels)), sub_labels, rotation=90); plt.yticks(range(len(sub_labels)), sub_labels)
    plt.tight_layout(); plt.savefig(out_png, dpi=160)
    if also_show:
        try: plt.show()
        except Exception: pass
    plt.close(); logging.info(f"[PLOT] Saved confusion heatmap → {out_png}")

# ---- Load sklearn artifacts ----
if not os.path.exists(PKL_PATH): raise FileNotFoundError(f"Missing sklearn pipeline PKL: {PKL_PATH}")
bundle=joblib.load(PKL_PATH); sk_pipe=bundle["pipeline"]; label_encoder=bundle["label_encoder"]
class_names=list(label_encoder.classes_); num_classes=len(class_names)
logging.info(f"[LOAD] Classes: {num_classes}")

# ---- Load Keras model (robust) ----
keras_model=None; keras_source=None
try:
    if os.path.exists(KERAS_PATH_ALT):
        # Preferred: SavedModel/.keras (Vectorizer state loads correctly)
        keras_model=load_model(KERAS_PATH_ALT, compile=False); keras_source=KERAS_PATH_ALT
    elif os.path.exists(H5_PATH):
        # H5 may fail due to TextVectorization table; we'll catch and continue
        keras_model=load_model(H5_PATH, compile=False); keras_source=H5_PATH
    else:
        logging.warning("[KERAS] No Keras model file found — will proceed with sklearn only.")
except Exception as e:
    logging.warning(f"[KERAS] Failed to load Keras model ({e}) — will proceed with sklearn only.")
    keras_model=None

# ---- Read input & build text ----
if not os.path.exists(INPUT_CSV): raise FileNotFoundError(f"Input CSV not found: {INPUT_CSV}")
logging.info(f"[READ] {INPUT_CSV}")
df=pd.read_csv(INPUT_CSV, engine="python"); df=normalize_cols(df)
text_cols=collect_text_columns(df); logging.info(f"[TEXT] Using columns: {text_cols}")
df["__text__"]=build_text_series(df, text_cols)
lbl_col=pick_label_column(df)
if lbl_col: df[lbl_col]=df[lbl_col].astype(str).map(canonicalize)
id_cols=[c for c in ["style","genre","artist","title"] if c in df.columns]
texts=df["__text__"].fillna("").astype(str).values

# ---- sklearn predictions ----
logging.info("[PRED] sklearn pipeline…")
prob_sk=None
try:
    prob_sk=sk_pipe.predict_proba(texts); idx_sk=np.argmax(prob_sk, axis=1); conf_sk=prob_sk[np.arange(len(idx_sk)), idx_sk]
except Exception:
    idx_sk=sk_pipe.predict(texts); conf_sk=np.full_like(idx_sk, np.nan, dtype=float)
pred_sk=label_encoder.inverse_transform(idx_sk)

# ---- keras predictions (safe try/except) ----
pred_k=None; conf_k=None; idx_k=None; prob_k=None
if keras_model is not None:
    try:
        logging.info(f"[PRED] Keras model… ({os.path.basename(keras_source)})")
        ds=tf.data.Dataset.from_tensor_slices(texts).batch(256)
        prob_k=keras_model.predict(ds, verbose=0)        # <-- may raise the LookupTable init error on H5
        idx_k=np.argmax(prob_k, axis=1); conf_k=prob_k[np.arange(len(idx_k)), idx_k]
        pred_k=label_encoder.inverse_transform(idx_k)
    except Exception as e:
        logging.warning(f"[PRED][KERAS] Skipping due to error: {e}")
        keras_model=None  # disable downstream heatmap use

# ---- Ensemble (if both probs available & aligned) ----
if prob_sk is not None and prob_k is not None and prob_sk.shape==prob_k.shape==(len(texts), num_classes):
    avg_prob=(prob_sk+prob_k)/2.0; idx_avg=np.argmax(avg_prob, axis=1); conf_avg=avg_prob[np.arange(len(idx_avg)), idx_avg]
    pred_avg=label_encoder.inverse_transform(idx_avg)
else:
    # fallback to sklearn-only
    idx_avg, conf_avg, pred_avg = idx_sk, conf_sk, pred_sk

# ---- Save predictions ----
out = pd.DataFrame({
    "pred_sklearn": pred_sk,
    "pred_sklearn_conf": conf_sk,
    "pred_avg": pred_avg,
    "pred_avg_conf": conf_avg
})
if pred_k is not None:
    out["pred_keras"]=pred_k
    out["pred_keras_conf"]=conf_k
if id_cols:
    out=pd.concat([df[id_cols].reset_index(drop=True), out.reset_index(drop=True)], axis=1)
os.makedirs(os.path.dirname(OUTPUT_CSV) or ".", exist_ok=True)
out.to_csv(OUTPUT_CSV, index=False, encoding="utf-8")
logging.info(f"[SAVE] Predictions → {OUTPUT_CSV}")

# ---- Plots ----
# Accuracy curve from training (if present)
if os.path.exists(HISTORY_CSV):
    def plot_accuracy_curve(history_csv, out_png, also_show=True):
        hist=pd.read_csv(history_csv); plt.figure(figsize=(9,6))
        if "accuracy" in hist.columns: plt.plot(hist["epoch"], hist["accuracy"], label="Train Accuracy")
        if "val_accuracy" in hist.columns: plt.plot(hist["epoch"], hist["val_accuracy"], label="Val Accuracy")
        if "top3_acc" in hist.columns: plt.plot(hist["epoch"], hist["top3_acc"], label="Train Top-3")
        if "val_top3_acc" in hist.columns: plt.plot(hist["epoch"], hist["val_top3_acc"], label="Val Top-3")
        plt.title("Training Accuracy"); plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend(); plt.grid(True, linestyle="--", linewidth=0.5)
        plt.tight_layout(); plt.savefig(out_png, dpi=160)
        try: plt.show()
        except Exception: pass
        plt.close(); logging.info(f"[PLOT] Saved accuracy curve → {out_png}")
    plot_accuracy_curve(HISTORY_CSV, ACC_PNG, also_show=True)
else:
    logging.warning(f"[PLOT] No history.csv at {HISTORY_CSV} — skipping accuracy plot.")

# Confusion heatmap if we have true labels AND a prediction index vector (use keras if available else sklearn)
if lbl_col:
    logging.info(f"[EVAL] Found ground-truth column '{lbl_col}' — building confusion heatmap.")
    y_true_names=df[lbl_col].values; is_known=np.isin(y_true_names, class_names)
    if np.any(is_known):
        y_true_idx=label_encoder.transform(y_true_names[is_known])
        idx_for_heatmap = idx_k if (keras_model is not None and idx_k is not None) else idx_sk
        y_pred_idx = idx_for_heatmap[is_known]
        # save full + top-25 heatmap
        from sklearn.metrics import confusion_matrix
        cm_full=confusion_matrix(y_true_idx, y_pred_idx, labels=np.arange(len(class_names)))
        pd.DataFrame(cm_full, index=class_names, columns=class_names).to_csv(CONF_FULL_CSV, encoding="utf-8")
        logging.info(f"[PLOT] Saved full confusion matrix CSV → {CONF_FULL_CSV}")
        true_counts=cm_full.sum(axis=1); top_idx=np.argsort(true_counts)[::-1][:min(25,len(class_names))]
        sub=cm_full[np.ix_(top_idx, top_idx)]; sub_labels=[class_names[i] for i in top_idx]
        plt.figure(figsize=(12,10)); im=plt.imshow(sub, interpolation="nearest", aspect="auto"); plt.colorbar(im, fraction=0.046, pad=0.04)
        plt.title(f"Confusion Matrix Heatmap (Top {len(sub_labels)} Classes)"); plt.xticks(range(len(sub_labels)), sub_labels, rotation=90); plt.yticks(range(len(sub_labels)), sub_labels)
        plt.tight_layout(); plt.savefig(HEATMAP_PNG, dpi=160)
        try: plt.show()
        except Exception: pass
        plt.close(); logging.info(f"[PLOT] Saved confusion heatmap → {HEATMAP_PNG}")
    else:
        logging.warning("[EVAL] None of the true labels match training classes — skipping heatmap.")
else:
    logging.info("[EVAL] No true label column — skipping confusion heatmap.")

print("Done.")
```
